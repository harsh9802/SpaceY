```python
import numpy as np
proximities = {
    'Miami': (25.7617, -80.1918),
    'Orlando': (28.5383, -81.3792),
    'Los Angeles': (34.0522, -118.2437),
    'San Francisco': (37.7749, -122.4194)
}

# Haversine formula to calculate the distance
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth radius in kilometers

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = R * c
    return distance

# Create a DataFrame to store distances
distances = pd.DataFrame(columns=['Launch Site'] + list(proximities.keys()))

# Calculate distances
for index, row in launch_sites_df.iterrows():
    distances.at[index, 'Launch Site'] = row['Launch Site']
    for city, coords in proximities.items():
        distance = haversine(row['Lat'], row['Long'], coords[0], coords[1])
        distances.at[index, city] = distance

# Display the distances DataFrame
print(distances)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
coastlines = [
    (28.56367, -80.57163), # Example coastline coordinate
    (29.63731, -95.20621), # Example coastline coordinate
    (34.00778, -118.49923), # Example coastline coordinate
    (37.81993, -122.47825)  # Example coastline coordinate
]
# Find the closest coastline coordinate for each launch site
closest_coastlines = []

for index, row in launch_sites_df.iterrows():
    min_distance = float('inf')
    closest_coord = None
    for coast in coastlines:
        distance = calculate_distance(row['Lat'], row['Long'], coast[0], coast[1])
        if distance < min_distance:
            min_distance = distance
            closest_coord = coast
    closest_coastlines.append({
        'Launch Site': row['Launch Site'],
        'Launch Site Lat': row['Lat'],
        'Launch Site Lon': row['Long'],
        'Closest Coastline Lat': closest_coord[0],
        'Closest Coastline Lon': closest_coord[1],
        'Distance to Coastline (km)': min_distance
    })

# Create a DataFrame for the closest coastlines
closest_coastlines_df = pd.DataFrame(closest_coastlines)

# Display the closest coastlines DataFrame
print(closest_coastlines_df)
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
cities = [(25.7617, -80.1918), (28.5383, -81.3792), (34.0522, -118.2437), (37.7749, -122.4194)]
railways = [(28.4024, -80.6453), (29.4241, -98.4936), (33.748995, -84.387982), (40.7128, -74.0060)]
highways = [(28.5383, -81.3792), (34.0522, -118.2437), (29.7604, -95.3698), (32.7767, -96.7970)]
```

```python
def add_proximity_markers(proximities, label):
    for index, row in launch_sites_df.iterrows():
        min_distance = float('inf')
        closest_coord = None
        for loc in proximities:
            distance = calculate_distance(row['Lat'], row['Long'], loc[0], loc[1])
            if distance < min_distance:
                min_distance = distance
                closest_coord = loc
        
        # Add marker for the closest proximity point
        folium.Marker(
            location=[closest_coord[0], closest_coord[1]],
            icon=folium.Icon(color='green', icon='info-sign'),
            popup=f'Closest {label}'
        ).add_to(site_map)
        
        # Add marker to display the distance
        folium.Marker(
            location=[(row['Lat'] + closest_coord[0]) / 2, (row['Long'] + closest_coord[1]) / 2],
            icon=DivIcon(
                icon_size=(20, 20),
                icon_anchor=(0, 0),
                html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:.2f} KM".format(min_distance),
            )
        ).add_to(site_map)
        
        # Draw a PolyLine between the launch site and the closest proximity point
        lines = folium.PolyLine(
            locations=[(row['Lat'], row['Long']), (closest_coord[0], closest_coord[1])],
            weight=1
        )
        site_map.add_child(lines)

# Add proximity markers for cities, railways, and highways
add_proximity_markers(cities, 'City')
add_proximity_markers(railways, 'Railway')
add_proximity_markers(highways, 'Highway')

# Display the map
site_map
```
